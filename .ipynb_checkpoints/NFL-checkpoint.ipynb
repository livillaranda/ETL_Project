```python
# Dependencies
import pandas as pd
```

```python
# Import AFC file
afcpath = 'NFL_AFC.xls'

afc = pd.read_excel(afcpath)
```

```python
# Create DataFrame
afc_df = pd.DataFrame(afc)
afc_df
```

```python
# Drop Unneeded Columns
afc_rec = afc_df[['Tm', 'W', 'L', 'W-L%']].copy()
afc_rec
```

```python
# Insert City Location
afc_cities = ['Boston', 'New York', 'Miami', 'Buffalo', 'Cincinnati', 'Baltimore', 'Pittsburgh', 'Cleveland', 'Indianapolis',
             'Houston', 'Nashville', 'Jacksonville', 'San Diego', 'Denver', 'Oakland', 'Kansas City']

afc_rec['City'] = afc_cities
```

```python
# Rename Column to 'Team Name'
afc_rec = afc_rec.rename(columns={'Tm':'Team Name'})
```

```python
# Verify Clean Data for AFC Record
afc_rec
```
